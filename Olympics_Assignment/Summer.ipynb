# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
print(len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
most=df[df['Medal'] == "Gold"]
most.groupby(['Sport'])['Medal'].count().reset_index().sort_values(by='Medal',ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[df['Sport']  == sport])])


pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (6,6))
```

### 4. Which player has won most number of medals? (Top 5)

```python
medal=df[['Athlete','Medal']]
medal.columns=[['Athlete','Medal']]
medals=df.groupby(['Athlete'])['Medal'].count().reset_index().sort_values(by='Medal',ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (5,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
medal2=df[df['Medal']=='Gold']
medal2.groupby(['Athlete'])['Medal'].count().reset_index().sort_values(by='Medal',ascending=False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (5,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for events in df['Event'].unique():
    data.append([events,len(df[df['Event']==events])])
    
pd.DataFrame(data,columns = ['Event','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Event', y = 'freq', kind = 'bar', figsize = (5,5))
    
```
